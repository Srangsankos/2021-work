```python
table_N = table.copy()
table_un = table.copy()
table_Volume = table.copy()
```

```python
#แถวที่1
for i in range(1,len(table_N.columns)):
    table_N.iloc[0,i] = 2500
    table_un.iloc[0,i] = 2500
    table_Volume.iloc[0,i] = table_N.iloc[0,i]/table_Open.iloc[0,i]
    

    
for i in range(1,len(table_N)):
    save_volume = {} #sace volume
    save_price = {} #save open price
    save_un = {}
    for j in range(1,len(table_N.columns)):
        save_price[name_table.iloc[i-1,j]] = table_N.iloc[i-1,j]
        save_volume[name_table.iloc[i-1,j]] = table_Volume.iloc[i-1,j]
        save_un[name_table.iloc[i-1,j]] = table_un.iloc[i-1,j]
        
    #print(save_price)
   
    name_old =[]     
    for k in range(1,11):
        if name_table.iloc[i-1,k] in name_table.iloc[i].to_list(): #check new row
            pass
        else:
            name_old.append(name_table.iloc[i-1,k])
            
    new_name = []
    for l in range(1,11):
        if name_table.iloc[i,l] in name_table.iloc[i-1].to_list():
            pass
        else:
            new_name.append(name_table.iloc[i,l])
            
    for m in range(len(name_old)):
        save_volume[new_name[m]] = save_volume.pop(name_old[m])
        save_price[new_name[m]] = save_price.pop(name_old[m])
        save_un[new_name[m]] = save_un.pop(name_old[m])
    #print(save_price)
   
    # zone cal
    for n in range(1,11):
        #ทุนที่เปิดposition
        
        if name_table.iloc[i,n] in name_table.iloc[i-1].to_list():
            table_Volume.iloc[i,n] = save_volume[name_table.iloc[i,n]]#table_N.iloc[i,n]/table_Open.iloc[i,n]
            table_un.iloc[i,n] = table_Volume.iloc[i,n] * table_n1.iloc[i,n]
            table_N.iloc[i,n] = save_price[name_table.iloc[i,n]]
            
        else:
            table_N.iloc[i,n] = save_un[name_table.iloc[i,n]]
            table_Volume.iloc[i,n] = table_N.iloc[i,n]/table_Open.iloc[i,n]
            table_un.iloc[i,n] = table_Volume.iloc[i,n] * table_n1.iloc[i,n]


table_Volume['Table'] = 'Volume'
table_Volume['N'] = 4

table_N['Table'] = 'Notional Cost ( Incl. Vat & Fee )'
table_N['N'] = 5

table_un['Table'] = 'unrealized P/L'
table_un['N'] = 6
```
