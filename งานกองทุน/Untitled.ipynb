```python
def read_data(country):
    data = pd.read_excel(r'/Users/oat/Desktop/งาน/data fund/'+country+'/Accum_week_flow.xlsx')
    return data
```

```python
def tran_data(country):
    data = read_data(country)
    data.set_index('Unnamed: 0',inplace=True)
    data = data.transpose()
    df = pd.DataFrame(index=data.index[4:],columns=range(1,6))
    for i in range(len(df)):
        dt_list = data.iloc[i,:].sort_values(ascending=False).dropna().index.to_list()
        for j in range(len(df.columns)):
            if not dt_list:
                pass
            else:
                df.iloc[i,j] = dt_list[j]
    df['table'] = country
    if country == 'US':
        df['N'] = 1
    elif country == 'Euro':
        df['N'] = 2
    elif country == 'Japan':
        df['N'] = 3
    elif country == 'India':
        df['N'] = 4
    return df
```
