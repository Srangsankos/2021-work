```python
def sheet(x):
    df[x] = df[x].sort_values('Time ').reset_index().drop(columns='index')
    
    IT = df[x].loc[0,'Close']
    
    for i in range(1,len(df[x])):
        if df[x].iloc[i,0].year != df[x].iloc[i-1,0].year: # หาวันที่เป็นตัวเทียบ
            IT = df[x].loc[i,'Close']     
        elif df[x].iloc[i,0].year == df[x].iloc[i-1,0].year:
            df[x].loc[i,'Change'] = df[x].loc[i,'Close']/IT
    
            
    df[x] = df[x].set_index('Time ')
       
    
    return df[x]
```
