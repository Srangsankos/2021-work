```python
def s(x,i):
    if x.loc[i,'Open'] < 2:
        #print(0.01)
        TP = x.loc[i,'Open']+(4*0.01)
        SL = x.loc[i,'Open']-(5*0.01)


    elif 2<= x.loc[i,'Open'] < 5:
        #print(0.02)
        TP = x.loc[i,'Open']+(4*0.02)
        SL = x.loc[i,'Open']-(5*0.02)


    elif 5<= x.loc[i,'Open'] < 10:
        #print(0.05)
        TP = x.loc[i,'Open']+(4*0.05)
        SL = x.loc[i,'Open']-(5*0.05)

    elif 10<= x.loc[i,'Open'] < 25:
        #print(0.1)
        TP = x.loc[i,'Open']+(4*0.1)
        SL = x.loc[i,'Open']-(5*0.1)


    elif 25<= x.loc[i,'Open'] < 100:
        #print(0.25)
        TP = x.loc[i,'Open']+(4*0.25)
        SL = x.loc[i,'Open']-(5*0.25)


    elif 100<= x.loc[i,'Open'] < 200:
        #print(0.5)
        TP = x.loc[i,'Open']+(4*0.5)
        SL = x.loc[i,'Open']-(5*0.5)


    elif 200<= x.loc[i,'Open'] < 400:
        #print(1)
        TP = x.loc[i,'Open']+(4)
        SL = x.loc[i,'Open']-(5)


    elif x.loc[i,'Open'] >= 400:
        #print(2)
        TP = x.loc[i,'Open']+(4*2)
        SL = x.loc[i,'Open']-(5*2)


    if x.loc[i,'High'] >= TP:
        Sell = TP
        
    elif x.loc[i,'Low'] <= SL:
        Sell = SL
        
    else:
        Sell = x.loc[i,'Close']
        
    x.loc[i,'TP'] = TP
    x.loc[i,'SL'] = SL
        
    return Sell
    
```
