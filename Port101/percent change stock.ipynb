```python
@lru_cache(maxsize=None)
def sheet(x):
    df[x] = df[x].sort_values('Time ').reset_index().drop(columns='index').copy()
    
    df3 = df[x].set_index('Time ').copy()
    df3.dropna(inplace=True)
    
    df2 = pd.DataFrame(columns=df3.columns,index=all_time)
    
    for t in range (len(df3.index)):
        if not pd.isna(df3['Close'][df3.index[t]]):
            for c in df2.columns:
                df2[c][df3.index[t]] = df3[c][df3.index[t]].copy()
    
    for t in range (len(df2)):
        if pd.isna(df2['Close'][df2.index[t]]):
            for c in df2.columns:
                df2[c][df2.index[t]] = df2[c][df2.index[t-1]]
    
    df2.dropna(inplace=True)
    df2 = df2.reset_index()
    
    IT = df2.loc[0,'Close']
    
    for i in range(1,len(df2)):
        if df2.iloc[i,0].year != df2.iloc[i-1,0].year: # หาวันที่เป็นตัวเทียบ
            IT = df2.loc[i,'Close']     
        elif df2.iloc[i,0].year == df2.iloc[i-1,0].year:
            df2.loc[i,'Change'] = df2.loc[i,'Close']/IT
    
            
    df2 = df2.set_index('index')
       
    
    return df2
```

```python
df_pctchange = pd.DataFrame(index=all_time)

for symbol in s100:
    df_pctchange[symbol] = np.nan
    for t in sheet(symbol).index:
        df_pctchange[symbol][t] = sheet(symbol)['Change'][t]

df_pctchange.head()
```

```python
signal_name_table = pd.DataFrame(index=all_time,columns=list(range(1,11)))
for t in range (1,len(all_time)):
    signal_name_table.iloc[t] = list(df_pctchange.iloc[t].sort_values(ascending=False)[:10].index)
    if signal_name_table.index[t].year != signal_name_table.index[t-1].year:
        signal_name_table.iloc[t] = signal_name_table.iloc[t-1]
signal_name_table['Table'] = 'Signal Name'
signal_name_table['N'] = 1
signal_name_table.head()
```
