```python
Time = df['Result-3-5'].loc[0,'Time ']
List = [df['Result-3-5'].loc[0,'%Change']]
for i in range(1,len(df['Result-3-5'])):
    if Time != df['Result-3-5'].loc[i,'Time ']:
        Time = df['Result-3-5'].loc[i,'Time ']
        df['Result-3-5'].loc[i-1,'%Change_end_day'] = sum(List)/len(List)
        List = []
        List.append(df['Result-3-5'].loc[i,'%Change'])

    elif Time == df['Result-3-5'].loc[i,'Time ']:
        List.append(df['Result-3-5'].loc[i,'%Change'])
        
        
```
