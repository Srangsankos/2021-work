```python
import pandas as pd
import codecs
import datetime
from datetime import date
import numpy as np
import os
```

```python
def getData(symbol,date):
    f = codecs.open(r'/Users/oat/Desktop/งาน/งานช้าง/'+symbol+'/'+date, "r", "utf-16")
    data = pd.read_csv(f ,sep = '\\t',skip_blank_lines=True)
#     data.reset_index(inplace = True)
#     list_col = data.columns[1:].to_list()
#     list_old_col = data.columns[:-1].to_list()
#     data.rename(columns={list_old_col[i]:list_col[i] for i in range(len(list_col))},inplace=True)
#     data = data.iloc[:,:-1].copy()
    status = 'close'
    for i in range(len(data)):
        if datetime.datetime.strptime(data.iloc[i,0], '%Y.%m.%d %H:%M:%f').strftime('%H:%M') == '16:30' or status == 'on':
            data.iloc[i,0] = np.nan
            status = 'on'
    data.dropna(inplace = True)
    return data
```

```python
def Condata(symbol):
    list_date = os.listdir('/Users/oat/Desktop/งาน/งานช้าง/'+symbol)
    list_date.sort()
    check = False
    i = 0
    while check == False:
        if list_date[i] == '2021.10.01.csv':
            R = list_date[i:].copy()
            check = True
        else:
            i += 1
            pass
    
    a = getData(symbol,R[0])
    for i in range(1,len(R)):        
        b = getData(symbol,R[i])
        a = pd.concat([a,b])

    return a
```

```python
def Caldata(symbol,date):
    data = getData(symbol,date)
    data['Sum_vol_bid'] = data['Vol Bid 1']+data['Vol Bid 2']+data['Vol Bid 3']+data['Vol Bid 4']+data['Vol Bid 5']
    data['Sum_vol_ask'] = data['Vol Ask 1']+data['Vol Ask 2']+data['Vol Ask 3']+data['Vol Ask 4']+data['Vol Ask 5']
    for i in range(1,len(data)):
        if data.loc[i,'Sum_vol_bid'] > data.loc[i,'Sum_vol_ask']:
            data.loc[i,'status'] = 'BUY'
#             print(i)
    return data
```
