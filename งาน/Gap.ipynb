```python
def sheet(x):
    df[x] = df[x].sort_values('Time ').reset_index().drop(columns='index')
    
    for i in range(1,len(df[x])):
        df[x].loc[i,'OI^'] = df[x].loc[i-1,'OI']
        df[x].loc[i,'%Change'] = ((df[x].loc[i,'Open']-df[x].loc[i-1,'Close'])/df[x].loc[i-1,'Close'])*100
    for i in range(1,len(df[x])):
        if df[x].loc[i,'OI^'] >= 10000000:
            pass
        else:
            df[x] = df[x].drop(i)
    
            
    df[x] = df[x].set_index('Time ')
        
    
    return df[x]
```
